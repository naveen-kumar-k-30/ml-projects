```python
df = pd.read_csv('demoData.csv')
df
```

```python
#to avoid duplicate values
df.drop_duplicates(inplace=True)
df
```

```python
#to replce col-name
df.rename(columns={'Name':'Full Name'},inplace=True)
df
```

```python
# to add new column
df['Gender']=['Male','Female','Female','Male','Male','Male','Male','Male']
df
```

```python
grp_df=df.groupby('Gender')['Age'].mean()
grp_df
```

```python
#to view the correlation
numeric_df= df.select_dtypes(include=[float,int])
corrdf = numeric_df.corr()
corrdf.corr()
```
