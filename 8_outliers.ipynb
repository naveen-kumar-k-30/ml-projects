```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Data array
data = np.array([250, 275, 280, 300, 320, 325, 330, 350, 360, 10000])

# Calculate Z-scores
z_scores = np.abs(stats.zscore(data))

# Define the threshold for identifying outliers
threshold = 3

# Find indices of outliers
outliers = np.where(z_scores > threshold)[0]
print("Outliers (indices):", outliers)

# Plotting
colors = "red"
plt.figure(figsize=(8, 6))
plt.scatter(np.arange(len(data)), data, c=colors)  # Corrected np.arrange to np.arange
plt.xlabel("Data Point Index")
plt.ylabel("Sale Price (in dollars)")
plt.title("Outliers Detection using Z-score")
plt.show()
```
